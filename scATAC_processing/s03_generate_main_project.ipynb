```r
library(ArchR)
library(parallel)
library(BSgenome.Mmusculus.UCSC.mm10)
library(ggpubr)
library(dplyr)

addArchRGenome("mm10")

set.seed(147)
```

# D30 Dataset: Generate scATAC Peak Set

```r
proj <- loadArchRProject('/data/peer/sotougl/Fuchs/inflammatory_memory/scATAC/intermediate_outputs3/projects/ArchR_D30_epdsc')
```

```r
proj@projectMetadata$outputDirectory = '/data/peer/sotougl/Fuchs/inflammatory_memory/scATAC/intermediate_outputs3/projects/ArchR_D30_min_NA_max_600_scpeakset'
proj@projectMetadata$outputDirectory
```

```r
# Generate group-level coverage tracks (fragments capped at 600bp for NFR enrichment)
proj <- addGroupCoverages(proj, maxFragmentLength=600, force = TRUE)
```

```r
# Call reproducible peaks across all groups using MACS3
proj <- addReproduciblePeakSet(proj, pathToMacs2  = '/home/sotougl/miniconda3/envs/archr_env/bin/macs3', force = TRUE, plot = FALSE, threads = 1)
```

```r
# Generate single-cell peak accessibility matrix (NFR fragments, no ceiling for high counts)
proj <- addPeakMatrix(proj, maxFragmentLength=600, ceiling=10^9)

# Persist project with peak annotations
proj <- saveArchRProject(ArchRProj = proj, outputDirectory = '/data/peer/sotougl/Fuchs/inflammatory_memory/scATAC/intermediate_outputs3/projects/ArchR_D30_min_NA_max_600_scpeakset')
```
